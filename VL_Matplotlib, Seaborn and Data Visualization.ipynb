```python
gas=pd.read_csv("C:\\Users\\user\\Desktop\\Data Science Bootcamp\\Dataset\\gas_prices.csv")
plt.figure(figsize=(8,5))

plt.title("Gas Prices Over Time (In USD)")

plt.xlabel("Year")
plt.ylabel("US Dollars")




plt.plot(gas.Year, gas.USA, c="b", label="USA", linewidth=4)
plt.plot(gas.Year, gas.Canada, c="r", label="Canada", linewidth=4)
plt.plot(gas.Year, gas["South Korea"], c="g", label="South Korea", linewidth=4)
plt.plot(gas.Year, gas.Australia, c="y", label="Australia", linewidth=4)
plt.legend()

print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.show()
```

#### Histogram

```python
fifa=pd.read_csv("C:\\Users\\user\\Desktop\\Data Science Bootcamp\\Dataset\\fifa_data.csv")
fifa.head(5)
```

```python
bins=[40, 50, 60, 70, 80, 90, 100]
plt.hist(fifa.Overall, bins=bins, color="green")
plt.xticks(bins)

plt.xlabel("Skill Level")
plt.ylabel("Number of Players")
plt.title("Distribution of Player Skills in FIFA 2018")
plt.show()
```

#### Pie Charts

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]

labels=["Left", "Right"]
colors=["#abcdef", "#aabbcc"]

plt.title("Foot Preference of FiFA players")


plt.pie([left, right], labels=labels, colors=colors)
plt.show()
```
